2. **¿Cómo se comportan las distribuciones anuales por trimestre?**

Vamos a calcular por cada año, cuales fueron los accesos a internet por cada 100 hogares por cada trimestre. 

```python
# funcion para calcular los totales de cada Trimestre y por Año.
def calcular_totales_trimestrales(anio, df, columna_a_evaluar):
    
    mask_t1_anio = (df['Año'] == anio) & (df['Trimestre'] == 1)
    mask_t2_anio = (df['Año'] == anio) & (df['Trimestre'] == 2)
    mask_t3_anio = (df['Año'] == anio) & (df['Trimestre'] == 3)
    mask_t4_anio = (df['Año'] == anio) & (df['Trimestre'] == 4)
    
    lista_totales_trimestrales = [df[mask_t1_anio][columna_a_evaluar].sum().round(2),df[mask_t2_anio][columna_a_evaluar].sum().round(2), df[mask_t3_anio][columna_a_evaluar].sum().round(2), df[mask_t4_anio][columna_a_evaluar].sum().round(2)]
    return lista_totales_trimestrales
```

```python
# cada variable es una lista con datos de cada uno de los trimestres del año.

A2022 = calcular_totales_trimestrales(2022, dt1, 'Accesos por cada 100 hogares')
A2021 = calcular_totales_trimestrales(2021, dt1, 'Accesos por cada 100 hogares')
A2020 = calcular_totales_trimestrales(2020, dt1, 'Accesos por cada 100 hogares')
A2019 = calcular_totales_trimestrales(2019, dt1, 'Accesos por cada 100 hogares')
A2018 = calcular_totales_trimestrales(2018, dt1, 'Accesos por cada 100 hogares')
A2017 = calcular_totales_trimestrales(2017, dt1, 'Accesos por cada 100 hogares')
A2016 = calcular_totales_trimestrales(2016, dt1, 'Accesos por cada 100 hogares')
A2015 = calcular_totales_trimestrales(2015, dt1, 'Accesos por cada 100 hogares')
A2014 = calcular_totales_trimestrales(2014, dt1, 'Accesos por cada 100 hogares')
```

3- **¿Cuáles fueron las 5 provincias con más accesos a internet por cada 100 hogares en el 2022?**

```python
def sumar_totales_por_provincia(data, year, orden):
    # Filtrar los datos para el año solicitado
    year_data = data[data['Año'] == year]
    
    # Sumar los totales por provincia
    totals = year_data.groupby('Provincia')['Accesos por cada 100 hogares'].sum().reset_index().round(2)
    
    # Ordenar los totales de forma descendente
    if orden =='ascendente':
        ascending = False
    else:
        ascending = True
    
    totals = totals.sort_values('Accesos por cada 100 hogares', ascending=ascending)
    
    # Crear una lista con los nombres de las provincias y los totales
    provincia_totals = totals[['Provincia', 'Accesos por cada 100 hogares']].values.tolist()
    
    return provincia_totals
```

```python
import matplotlib.pyplot as plt 


def crecimiento_anual(lista_totales, anio_inicio, anio_termino):
    # Crear una lista de años
    years = range(anio_inicio, anio_termino)

    # Crear una lista de los totales por año
    totals= lista_totales

    # Calcular el cambio porcentual entre años consecutivos
    changes = [(totals[i] - totals[i-1]) / totals[i-1] * 100 for i in range(1, len(totals))]

    # Crear una lista de colores para las barras
    colors = ['red' if change < 0 else 'green' for change in changes]

    # Crear el gráfico de barras con los colores correspondientes
    plt.bar(years[1:], changes, color=colors)

    # Agregar etiquetas y título al gráfico
    plt.xlabel('Año')
    plt.ylabel('Cambio porcentual')
    plt.title('Crecimiento/Decrecimiento respecto al año anterio (Banda Ancha)')

    # Agregar los valores porcentuales a las barras
    for i, change in enumerate(changes):
        plt.text(years[i+1], change, f'{change:.2f}%', ha='center', va='bottom')

    # Mostrar el gráfico
    plt.show()
```

**2. ¿Cómo se comportan las distribuciones anuales por Trimestre de Banda Ancha y Dial Up?**

```python
def calcular_totales_trimestrales(anio, df, columna):
    
    mask_t1_anio = (df['Año'] == anio) & (df['Trimestre'] == 1)
    mask_t2_anio = (df['Año'] == anio) & (df['Trimestre'] == 2)
    mask_t3_anio = (df['Año'] == anio) & (df['Trimestre'] == 3)
    mask_t4_anio = (df['Año'] == anio) & (df['Trimestre'] == 4)
    
    lista_totales_trimestrales = [df[mask_t1_anio][columna].sum().round(2),df[mask_t2_anio][columna].sum().round(2), df[mask_t3_anio][columna].sum().round(2), df[mask_t4_anio][columna].sum().round(2)]
    return lista_totales_trimestrales
```

```python

def graficar_lineas_trimestres (lista_totales, eje_xlabel, eje_ylabel, title_label):

    # Crear listas de totales por trimestre para cada año
    totals_trimestres = lista_totales

    # Crear una lista de trimestres
    trimestres = ['Trimestre 1', 'Trimestre 2', 'Trimestre 3', 'Trimestre 4']

    # Crear el gráfico de líneas
    for i, totals in enumerate(totals_trimestres):
    # Verificar si el año es 2022 y el último valor es cero
        if i == len(totals_trimestres) - 1 and totals[-1] == 0:
            totals = totals[:-1]  # Excluir el último valor
        plt.plot(trimestres[:len(totals)], totals, marker='o', label=f'Año {2014 + i}')

    # Agregar etiquetas y título al gráfico
    plt.xlabel(eje_xlabel)
    plt.ylabel(eje_ylabel)
    plt.title(title_label)

    # Mover la leyenda al costado fuera del gráfico
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    # Mostrar el gráfico
    plt.show()
```

```python
def sumar_totales_por_provincia(data, year, orden, feature):
    # Filtrar los datos para el año solicitado
    year_data = data[data['Año'] == year]
    
    # Sumar los totales por provincia
    totals = year_data.groupby('Provincia')[feature].sum().reset_index().round(2)
    
    # Ordenar los totales de forma descendente
    if orden =='mayor_a_menor':
        ascending = False
    else:
        ascending = True
    
    totals = totals.sort_values(feature, ascending=ascending)
    
    # Crear una lista con los nombres de las provincias y los totales
    provincia_totals = totals[['Provincia', feature]].values.tolist()
    
    return provincia_totals
```
